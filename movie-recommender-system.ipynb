# convert string of dictionary into list

```python
import ast  # Used to safely convert text into Python data
```

```python
def convert(text):
    L = []  # Create an empty list to store names

    # Convert text into a Python list and loop through it
    for i in ast.literal_eval(text): # Get the value of 'name' from each dictionary
        L.append(i['name'])  

    # Return the list of names
    return L
```

# update columns dictionary to list

```python
movies["genres"]=movies["genres"].apply(convert)
movies.head()
```

```python
movies["keywords"]=movies["keywords"].apply(convert)
movies.head()
```

```python
#get first three name of actors in cast
def convert3(obj):
    l=[]
def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L 
```

```python
movies["cast"]=movies["cast"].apply(convert3)
movies.head(2)
```

# Fetch director from crew

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
movies["crew"]=movies["crew"].apply(fetch_director)
movies.head(2)
```

```python
#remove space between name
def collapse(txt):
    l=[]
    for i in txt:
        l.append(i.replace(" ",""))
    return l
        
```

```python
movies["genres"]=movies["genres"].apply(collapse)
movies["keywords"]=movies["keywords"].apply(collapse)
movies["cast"]=movies["cast"].apply(collapse)
movies["crew"]=movies["crew"].apply(collapse)
movies.head(2)
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
new = movies.drop(columns=['overview','genres','keywords','cast','crew'])
#new.head()
```

```python
new['tags'] = new['tags'].apply(lambda x: " ".join(x))
new.head()
```

# Recommand movies on the basis of tags

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
    
```

```python
vector = cv.fit_transform(new['tags']).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
new[new['title'] == 'The Lego Movie'].index[0]
```

```python
def recommend(movie):
    index = new[new['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new.iloc[i[0]].title)
        
```

```python
recommend('Gandhi')
```
